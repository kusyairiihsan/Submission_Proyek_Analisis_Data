## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah ada hubungan antara cuaca dengan jumlah sewa sepeda?

```python
plt.figure(figsize=(8,4))
sns.barplot(
    x='weathersit',
    y='cnt',
    data=bike_day_df,
    hue='weathersit',
    palette='plasma',
    legend=False)

plt.title('Jumlah Penyewa Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Keterangan: 1 = Cerah, 2 = Mendung, 3 = Hujan')
plt.ylabel('Jumlah Penyewa Sepeda')
plt.show()
```

### Pertanyaan 2: Bagaimana pattern penggunaan sepeda berubah dari waktu ke waktu?

```python
plt.figure(figsize=(10,5))
bike_day_df['dteday'] = pd.to_datetime(bike_day_df['dteday'])
plt.plot(
    bike_day_df['dteday'],
    bike_day_df['cnt'],
    marker='o',
    linestyle='-')

plt.title('pattern penggunaan sepeda berubah dari waktu ke waktu')
plt.xlabel('Waktu')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.grid(True)
plt.show()
```

## Conclusion

- Conclution pertanyaan 1 : Dilihat dari output diatas menunjukkan bahwa cuaca mempengaruhi jumlah penyewa sepeda dikarenakan adanya penurunan jumlah penyewa dari cerah hingga ke hujan, pada hari cerah dan mendung dapat terlihat jumlah persewaan melebihi 4000, sedangkan pada cuaca hujan jumlah persewaan berkurang hingga hanya dibawah 2000.

- Conclution pertanyaan 2 : Dilihat dari grafik pattern penggunaan sepeda dari waktu ke waktu menunjukkan bahwa jumlah sepeda yang disewa cenderung meningkat pada beberapa bulan pertama dan menurun pada akhir tahun, hal ini mungkin terkait dengan perubahan musim.
